Create mock data


```python
data = {
    "cpu_temperature": [23.8, 24.1, 25.0, 22.5, 21.9],
    "cpu_temperature_zscore": [-0.5, 0.1, 1.2, -1.0, -1.8],
}
df = pd.DataFrame(data)

print(df)
```

```python
# List all column names
print(df.columns)

# Ensure you're using the correct names for computations
numeric_columns = ["cpu_temperature"]  # Adjust based on your dataset
for col in numeric_columns:
    if col not in df.columns:
        print(f"Column '{col}' not found!")
```

Visualize the Data


```python
import matplotlib.pyplot as plt

# Plot all numeric columns
numeric_columns = df.select_dtypes(include=['float64', 'int64']).columns  # Automatically detect numeric columns

# Plot each column
for col in numeric_columns:
    plt.figure(figsize=(10, 5))
    plt.plot(df[col], label=col, marker='o')
    plt.title(f"{col} Over Time")
    plt.xlabel("Index")
    plt.ylabel(col)
    plt.legend()
    plt.grid(True)
    plt.show()
```

 Detect Anomalies

```python
import numpy as np

# Compute Z-scores for numeric columns
z_scores = (df[numeric_columns] - df[numeric_columns].mean()) / df[numeric_columns].std()

# Add Z-score columns to DataFrame
for col in numeric_columns:
    df[f"{col}_zscore"] = z_scores[col]

# Identify anomalies (absolute Z-score > 3)
anomaly_threshold = 3
for col in numeric_columns:
    anomaly_col = f"{col}_anomaly"
    df[anomaly_col] = (df[f"{col}_zscore"].abs() > anomaly_threshold).astype(int)  # 1 = anomaly, 0 = normal

# Display anomaly summary
print("\nAnomaly Counts by Column:")
for col in numeric_columns:
    anomaly_col = f"{col}_anomaly"
    print(f"{col}: {df[anomaly_col].sum()} anomalies detected")
```

Visualize Anomalies

```python
# Plot anomalies for each column
for col in numeric_columns:
    plt.figure(figsize=(10, 5))
    plt.plot(df[col], label="Normal", color="blue", marker='o')

    # Highlight anomalies
    anomalies = df[df[f"{col}_anomaly"] == 1]
    plt.scatter(anomalies.index, anomalies[col], color="red", label="Anomaly", zorder=5)

    plt.title(f"{col} with Anomalies")
    plt.xlabel("Index")
    plt.ylabel(col)
    plt.legend()
    plt.grid(True)
    plt.show()
```

Save Results

```python
output_file = "synthetic_dataset_with_anomalies.csv"
df.to_csv(output_file, index=False)
print(f"Processed data saved to {output_file}")
```
